```python
 ## Dependent and independent feature

x = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
x
```

```python
y
```

```python
## Train Test Split
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=42)
```

```python
## Standardize the dataset
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
```

```python
x_train = scaler.fit_transform(x_train)
```

```python
x_test = scaler.transform(x_test)
```

```python
x_train
```

```python
x_test
```

```python
## Model Training 
from sklearn.linear_model import LinearRegression 
```

```python
regresor = LinearRegression()
```

```python
regresor.fit(x_train, y_train)
```

```python
## Print the coefficients and the intercept
print(regresor.coef_)
```

```python
print(regresor.intercept_)
```

```python
## on which parameters the model has been trained 
regresor.get_params()
```

```python
## Prediction with test Data 
reg_pred = regresor.predict(x_test)
```

```python
reg_pred
```

```python
plt.scatter(y_test, reg_pred)
plt.show()
```

```python
## Residuals 
resi = y_test-reg_pred
resi
```

```python
## Plot this residuals
sns.displot(resi, kind="kde")
plt.show()
```

```python
## Scatter plot with respect to prediction and residuals 
## Uniform distribution
plt.scatter(reg_pred, resi)
plt.show()
```

```python
from sklearn.metrics import mean_squared_error 
from sklearn.metrics import mean_absolute_error

print(mean_absolute_error(y_test,reg_pred))
print(mean_squared_error(y_test, reg_pred))
print(np.sqrt(mean_squared_error(y_test, reg_pred)))
```

```python
from sklearn.metrics import r2_score
score = r2_score(y_test,reg_pred)
print(score)
```

```python
## Adjusted r2
1 - (1- score)*(len(y_test)-1)/(len(y_test)-x_test.shape[1]-1)
```

```python
y_train_pred = regresor.predict(x_train)
mse_train = mean_squared_error(y_train, y_train_pred)
print(f"Mean Squared Error on Training Set: {mse_train}")
```
